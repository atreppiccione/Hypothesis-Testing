```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
from scipy.stats import stats
import re
```

```python
#open the file
fhandle = 'university_towns.txt'
univ = open(fhandle)
data = []

#parse each row and append state and RegioName to the list data
for row in univ:
     if '[edit]' in row:
        state = row[:re.search('\[', row).start()]
     else:
        if ' (' in row:
            RegionName = row[:re.search(' \(',row).start()] 
            data.append([state, RegionName])
        elif ' (' not in row:
            RegionName = row.rstrip()
            data.append([state, RegionName])


def get_list_of_university_towns():
     univ = pd.DataFrame(data, columns = ["State", "RegionName"])
     #st = pd.Series(states)
     #st = st.to_frame(name=None)
     #st.columns = (['State'])  
     #univ_new = pd.merge(univ, st, how = 'left', left_index = True, right_on = 'State')
     return univ

get_list_of_university_towns()
```
